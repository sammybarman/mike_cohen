```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
N = 10000

#initialize x and y vectors
sx = np.zeros(N)
sy = np.zeros(N)

for i in range(1,N):
    #generate a random number from {1,2,3}
    k = random.randint(1,3)

    #calculate the new point
    sx[i] = sx[i-1]/2 + (k-1)
    sy[i] = sy[i-1]/2

    if k ==2:
        sy[i] += 2

#plot the points
plt.plot(sx,sy,'k.',markersize=1)
plt.title('Sierpinski Triangle')
plt.axis('off')
plt.show()
```
